# # Sample data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Sample data
data = {
    'age': [25, 30, 35, 40, 45],
    'income': [50000, 60000, 70000, 80000, 90000]
}
df = pd.DataFrame(data)

# Features and target variable
x = df[['age']]  # Reshape to 2D array
y = df['income']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Optionally, evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred) #  - R² Score: معیاری برای توضیح واریانس است و به ما می‌گوید که چقدر از تغییرات هدف توسط مدل توضیح داده می‌شود.

# Print metrics
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)

# Plotting
plt.scatter(x, y, color='green', label="real data")

# Create a range of values for the regression line
x_range = np.linspace(x['age'].min(), x['age'].max()).reshape(-1, 1)
y_range = model.predict(x_range)

plt.plot(x_range, y_range, color='red', label="linear regression")
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Linear Regression")
plt.legend()
plt.show()
```
