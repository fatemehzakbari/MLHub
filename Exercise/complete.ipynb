# # Sample data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Sample data
data = {
    'age': [25, 30, 35, 40, 45],
    'income': [50000, 60000, 70000, 80000, 90000]
}
df = pd.DataFrame(data)

# Features and target variable
x = df[['age']]  # Reshape to 2D array
y = df['income']

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Optionally, evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred) #  - R² Score: معیاری برای توضیح واریانس است و به ما می‌گوید که چقدر از تغییرات هدف توسط مدل توضیح داده می‌شود.

# Print metrics
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)

# Plotting
plt.scatter(x, y, color='green', label="real data")

# Create a range of values for the regression line
x_range = np.linspace(x['age'].min(), x['age'].max()).reshape(-1, 1)
y_range = model.predict(x_range)

plt.plot(x_range, y_range, color='red', label="linear regression")
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Linear Regression")
plt.legend()
plt.show()
```

# DecisionTreeRegressor


**DecisionTreeRegressor :**

از دسته الگوریتم‌های یادگیری ماشین و به‌خصوص رگرسیون است

این الگوریتم به‌عنوان یک مدل درخت تصمیم عمل می‌کند که برای پیش‌بینی مقادیر پیوسته استفاده می‌شود

**ویژگی‌های کلیدی :**

قابلیت تفسیر: ساختار درختی آن باعث می‌شود که مدل به‌راحتی قابل تفسیر باشد.

غیرخطی: می‌تواند روابط غیرخطی بین ویژگی‌ها و هدف را مدل‌سازی کند.

عدم نیاز به پیش‌پردازش زیاد: معمولاً نیاز به نرمال‌سازی یا مقیاس‌بندی داده‌ها ندارد.


```python
ex = np.array([0,1,2,3,4,6,8,10]).reshape(-1, 1)
sa = np.array([30000,40000,55000,60000,70000,80000,85000,87000])
model = DecisionTreeRegressor(max_depth=2)
model.fit(ex, sa)
sa_predict = model.predict(ex)
plt.plot(ex, sa, marker="*", color="green")
plt.plot(ex, sa_predict,color="red")
sa_predict = model.predict([[11]])
print(sa_predict)
plt.xlabel("experienc")
plt.ylabel("salary")
plt.show()
```

# Random Forest


![My Image](/home/siyamak/Documents/MLHub/images/An-example-of-random-forest-classifier.png)
الگوریتم رندوم فارست (Random Forest) یک روش یادگیری ماشین مبتنی بر درخت تصمیم است که برای مسائل طبقه‌بندی و رگرسیون استفاده می‌شود. این الگوریتم به طور همزمان چندین درخت تصمیم را آموزش می‌دهد و نتایج آن‌ها را ترکیب می‌کند تا پیش‌بینی نهایی را به دست آورد.
ویژگی‌های کلیدی رندوم فارست:


جمع‌آوری درخت‌ها: رندوم فارست شامل مجموعه‌ای از درخت‌های تصمیم است که به صورت تصادفی از زیرمجموعه‌ای از داده‌ها و ویژگی‌ها آموزش می‌بینند. این کار به کاهش واریانس و افزایش دقت کمک می‌کند.


روش Bootstrap: برای آموزش هر درخت، از تکنیک نمونه‌برداری با جایگزینی (bootstrap sampling) استفاده می‌شود، که به هر درخت اجازه می‌دهد تا بر روی نمونه‌های مختلفی از داده‌ها آموزش ببیند.


انتخاب تصادفی ویژگی‌ها: در هر گره از درخت، به طور تصادفی تعدادی از ویژگی‌ها انتخاب می‌شوند تا بهترین تقسیم را انجام دهند. این کار به تنوع درخت‌ها کمک می‌کند.


رای‌گیری یا میانگین‌گیری: برای مسائل طبقه‌بندی، رندوم فارست با استفاده از روش رای‌گیری (majority voting) پیش‌بینی نهایی را تعیین می‌کند. برای مسائل رگرسیون، میانگین پیش‌بینی‌های درخت‌ها محاسبه می‌شود.


مزایا:

دقت بالا: به دلیل ترکیب چندین درخت، دقت پیش‌بینی معمولاً بالاست.
مقابله با بیش‌برازش: رندوم فارست به دلیل تنوع درخت‌ها، کمتر دچار بیش‌برازش می‌شود.
قابلیت تفسیر: می‌توان اهمیت ویژگی‌ها را با استفاده از تکنیک‌های مختلف در رندوم فارست اندازه‌گیری کرد.

معایب:

پیچیدگی محاسباتی: رندوم فارست ممکن است زمان‌بر باشد و نیاز به منابع محاسباتی بیشتری داشته باشد.
عدم تفسیرپذیری: اگرچه می‌توان اهمیت ویژگی‌ها را بررسی کرد، اما تفسیر کلی مدل نسبت به یک درخت تصمیم ساده‌تر دشوارتر است.

نتیجه‌گیری:
رندوم فارست یک الگوریتم قدرتمند و انعطاف‌پذیر است که در بسیاری از کاربردهای یادگیری ماشین به دلیل دقت و توانایی مقابله با داده‌های بزرگ و پیچیده بسیار مورد استفاده قرار می‌گیرد.
